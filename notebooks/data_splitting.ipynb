# Manual

```python
import os
import json
from sklearn.model_selection import train_test_split

manual_path = "../data/labels_manual"
automatic_path = "../data/inference/random_n3_scn3_pt_ns.jsonl"

data_path = "../data/lyrics"

template = {
    "id": None,
    "context": None,
    "entities": [
        {
            "text": None,
            "answer_start": None
        }
    ]
}
```

```python
import os
import json

full_data = []

label_path = manual_path

for filename in os.listdir(label_path):
    if filename.endswith(".json") and "template" not in filename:
        with open(os.path.join(label_path, filename), "r") as f:
            labels = json.load(f)
        
        with open(os.path.join(data_path, filename.replace(".json", ".txt")), "r") as f:
            context = f.read()
        
        data_entry = template.copy()
        data_entry["id"] = filename.replace(".json", "")
        data_entry["context"] = context
        data_entry["entities"] = labels
        
        full_data.append(data_entry)
        
len(full_data)
```

```python
import sys
import os
import json

automatic_data = []

with open(automatic_path, "r") as f:
    for line in f:
        data_entry = json.loads(line)
        
        filename = os.path.join(data_path, "{}.txt".format(data_entry["id"]))
        if not os.path.exists(filename):
            print("File not found:", filename, file=sys.stderr)
            continue
        
        with open(filename, "r") as f:
            context = f.read()
        data_entry["context"] = context
        
        automatic_data.append(data_entry)

len(automatic_data)
```

```python
from sklearn.model_selection import train_test_split

train, dev_test = train_test_split(full_data, test_size=0.3, random_state=42)
dev, test = train_test_split(dev_test, test_size=0.5, random_state=42)
```

```python
save_path = "../data/extraction/manual"

os.makedirs(save_path, exist_ok=True)

with open(os.path.join(save_path, "train_full.json"), "w") as f:
    json.dump(train+dev+test, f, indent=4)
    
with open(os.path.join(save_path, "train.json"), "w") as f:
    json.dump(train, f, indent=4)

with open(os.path.join(save_path, "dev.json"), "w") as f:
    json.dump(dev, f, indent=4)

with open(os.path.join(save_path, "test.json"), "w") as f:
    json.dump(test, f, indent=4)
```

# Mixed

```python
from sklearn.model_selection import train_test_split

train, dev_test = train_test_split(full_data, test_size=0.3, random_state=42)
dev, test = train_test_split(dev_test, test_size=0.5, random_state=42)
```

```python
len(train), len(dev), len(test)
```

```python
def mix(manual_data, automatic_data, ratio):
    train, dev_test = train_test_split(manual_data, test_size=0.3, random_state=42)
    dev, test = train_test_split(dev_test, test_size=0.5, random_state=42)
    
    train += automatic_data[:int(ratio*len(train))]
    
    print(len(train), len(dev), len(test))
    
    return train, dev, test
```

```python
for ratio in [0.3, 0.5, 1, 1.5, 2]:
    save_path = f"../data/extraction/mixed_{ratio}"
    os.makedirs(save_path, exist_ok=True)
    
    train, dev, test = mix(full_data, automatic_data, ratio=ratio)

    with open(os.path.join(save_path, "train_full.json"), "w") as f:
        json.dump(train+dev+test, f, indent=4)
        
    with open(os.path.join(save_path, "train.json"), "w") as f:
        json.dump(train, f, indent=4)

    with open(os.path.join(save_path, "dev.json"), "w") as f:
        json.dump(dev, f, indent=4)

    with open(os.path.join(save_path, "test.json"), "w") as f:
        json.dump(test, f, indent=4)
```
